# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. **Prompt1**:
Develop a advance Python program to manage inventory records for a prodcut managing company.
The system needs functions for adding, deleting and changing products and basic search options.
Your system should use CSV files to automate restocking and produce reports that show item values
alongside category breakdowns.
Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os
import random

# Constants
CSV_FILE = "inventory.csv"
RESTOCK_THRESHOLD = 5  # If stock < 5, restock automatically
RESTOCK_AMOUNT = 20  # Amount to restock
CATEGORIES = ["Electronics", "Clothing", "Groceries", "Home", "Toys"]

# Random Product Names
PRODUCT_NAMES = {
    "Electronics": ["Laptop", "Smartphone", "Tablet", "Headphones", "Monitor"],
    "Clothing": ["T-Shirt", "Jeans", "Jacket", "Sneakers", "Cap"],
    "Groceries": ["Milk", "Bread", "Eggs", "Rice", "Cereal"],
    "Home": ["Sofa", "Table", "Chair", "Lamp", "Curtains"],
    "Toys": ["Action Figure", "Board Game", "Puzzle", "Doll", "RC Car"]
}

# Generate Random Products
def generate_random_products(num_products=50):

    '''
    Creates a Products list and generates the random data for the inventory using random module

    Args: we need to give the number of records to be generated as an argument
    '''
    products = []
    for i in range(num_products):
        category = random.choice(CATEGORIES)
        name = random.choice(PRODUCT_NAMES[category])
        product_id = random.randint(1000, 9999)
        stock = random.randint(1, 30)
        price = round(random.uniform(5, 1000), 2)  # Prices between $5 and $1000
        products.append([product_id, name, category, stock, price])
    return products

# Initialize CSV with random data
def initialize_inventory():

    '''
    Checks if the inventory CSV file exists or not. If the CSV file doesn't exists then it calls the function
    generate_random_products() to generate the random data.

    The random generated data will be writed in to the CSV file. Thus, this function initializes the inventory file.
    '''
    if not os.path.exists(CSV_FILE):
        products = generate_random_products()
        with open(CSV_FILE, "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["ID", "Name", "Category", "Stock", "Price"])
            writer.writerows(products)
        print("Inventory initialized with randomly generated data.")

# Load inventory from CSV
def load_inventory():

    '''
    This loads the CSV data in to a list of dictionaries to perform other operations.

    If any exception occurs while reading the file like file is corrupted or unreadable file
    then it automatically created a new file calling initialize_inventory()
    '''
    try:
        with open(CSV_FILE, "r") as file:
            reader = csv.DictReader(file)
            return list(reader)
    except Exception:
        print("Error reading inventory file. Creating a new one.")
        initialize_inventory()
        return load_inventory()

# Save inventory back to CSV
def save_inventory(inventory):

    '''
    Once all the changes like modifying, adding or deleting any product information,
    these changes will be tracked in list of dictionaries that was created
    So, to reflec those changes in the origial CSV file this fuction overwrites the data in to CSV file.
    '''

    with open(CSV_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["ID", "Name", "Category", "Stock", "Price"])
        for item in inventory:
            writer.writerow([item["ID"], item["Name"], item["Category"], item["Stock"], item["Price"]])

# Add new product
def add_product():

    '''
    If we want to add any new product to our inventory then we can do by using this function.
    it takes runtme inputs for the product details and adds the product in to inventory.

    Raises:
    ValueError: when invalid input is given by user like entering string literal for numerical values.
    '''

    try:
        inventory = load_inventory()
        new_id = int(input("Enter Product ID: "))
        name = input("Enter Product Name: ")
        category = input(f"Enter Category {CATEGORIES}: ")
        stock = int(input("Enter Stock Quantity: "))
        price = float(input("Enter Price: "))

        if category not in CATEGORIES:
            print("Invalid category. Product not added.")
            return

        inventory.append({"ID": new_id, "Name": name, "Category": category, "Stock": stock, "Price": price})
        save_inventory(inventory)
        print(f"Product '{name}' added successfully!")
    except ValueError:
        print("Invalid input! Please enter correct values.")


# Delete product
def delete_product():

    '''
    This function takes the produc id that has to be deleted durig runtime from user and removes the product from inventory.

    If the product id given id not found then it just prints "Product ID not found"

    '''

    inventory = load_inventory()
    product_id = input("Enter Product ID to delete: ")

    new_inventory = [item for item in inventory if item["ID"] != product_id]

    if len(new_inventory) == len(inventory):
        print("Product ID not found.")
    else:
        save_inventory(new_inventory)
        print("Product deleted successfully.")

# Modify product
def modify_product():

    '''
    This is used to modify any details of the product and save the inventory in to a CSV file
    '''
    inventory = load_inventory()
    product_id = input("Enter Product ID to modify: ")

    for item in inventory:
        if item["ID"] == product_id:
            item["Name"] = input("Enter new Name: ") or item["Name"]
            item["Category"] = input("Enter new Category: ") or item["Category"]
            item["Stock"] = input("Enter new Stock: ") or item["Stock"]
            item["Price"] = input("Enter new Price: ") or item["Price"]
            save_inventory(inventory)
            print("Product modified successfully.")
            return

    print("Product ID not found.")

# Search product
def search_product():

    '''This fundtion is used to search for any products. The search should be performed irrespective of alphabetical case.
    So, the input as well as the names in file are converted in to lower case while comparision.
    '''

    inventory = load_inventory()
    search_term = input("Enter product name or category: ").lower()
    results = [item for item in inventory if search_term in item["Name"].lower() or search_term in item["Category"].lower()]

    if results:
        print("Search Results:")
        for item in results:
            print(item)
    else:
        print("No matching products found.")

# Restock items automatically
def auto_restock():
    '''
    It checks if any of the product in the inventory is below threshold level.
    If yes then that product will be restocked automatically
    '''
    inventory = load_inventory()
    restocked_items = []

    for item in inventory:
        if int(item["Stock"]) < RESTOCK_THRESHOLD:
            item["Stock"] = str(int(item["Stock"]) + RESTOCK_AMOUNT)
            restocked_items.append(item["Name"])

    save_inventory(inventory)

    if restocked_items:
        print("Restocked:", ", ".join(restocked_items))
    else:
        print("No restocking needed.")

# Generate report
def generate_report():

    '''
    This generated the total and categorical asset report for the inventory.
    it products the quantity and price to calculates the total stock value we have.
    '''

    inventory = load_inventory()
    category_values = {}

    total_value = 0
    for item in inventory:
        value = int(item["Stock"]) * float(item["Price"])
        total_value += value
        category_values[item["Category"]] = category_values.get(item["Category"], 0) + value

    print("\n=== Inventory Report ===")
    print(f"Total Inventory Value: ${total_value:.2f}")
    print("Category Breakdown:")
    for category, value in category_values.items():
        print(f"{category}: ${value:.2f}")

# Main menu
def main():
    '''
    In the main function, a sequnce of numbers will  be shown to select ay operation
    based on the number selection, the operation will be done to the inventory
    '''
    initialize_inventory()

    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Modify Product")
        print("4. Search Product")
        print("5. Restock Items")
        print("6. Generate Report")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_product()
        elif choice == "2":
            delete_product()
        elif choice == "3":
            modify_product()
        elif choice == "4":
            search_product()
        elif choice == "5":
            auto_restock()
        elif choice == "6":
            generate_report()
        elif choice == "7":
            print("Exiting... Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
import random
import csv
import os

# Constants
CSV_FILE = "flights.csv"
SEAT_CAPACITY = 100  # Each flight has 100 seats
FLIGHT_COUNT = 50

# Sample Cities for Flights
CITIES = ["New York", "Los Angeles", "Chicago", "Dallas", "San Francisco",
          "Miami", "Seattle", "Denver", "Boston", "Houston"]

# Generate random flights
def generate_flights(num_flights=FLIGHT_COUNT):
    '''
    This function is used to generate the list of flights that are available.
    The data is being generated using random module
    It takes number as an argument that tells the number of flights reuired in initial dataset
    '''
    flights = []
    for _ in range(num_flights):
        flight_id = random.randint(1000, 9999)  # Unique Flight ID
        departure = random.choice(CITIES)
        destination = random.choice([city for city in CITIES if city != departure])  # Different city
        price = round(random.uniform(100, 1000), 2)  # Ticket price range
        available_seats = SEAT_CAPACITY  # Initial seat availability
        flights.append([flight_id, departure, destination, price, available_seats])
    return flights

# Initialize flight data
def initialize_flight_system():
    if not os.path.exists(CSV_FILE):
        flights = generate_flights()
        with open(CSV_FILE, "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["Flight ID", "Departure", "Destination", "Price", "Available Seats"])
            writer.writerows(flights)
        print("Flight system initialized with 50 random flights.")

# Load flights from CSV
def load_flights():
    try:
        with open(CSV_FILE, "r") as file:
            reader = csv.DictReader(file)
            return list(reader)
    except Exception:
        print("Error reading flight file. Initializing new flights.")
        initialize_flight_system()
        return load_flights()

# Save flights back to CSV
def save_flights(flights):
    with open(CSV_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Flight ID", "Departure", "Destination", "Price", "Available Seats"])
        for flight in flights:
            writer.writerow([flight["Flight ID"], flight["Departure"], flight["Destination"],
                             flight["Price"], flight["Available Seats"]])

# Add a new flight
def add_flight():
    '''
    This function asks for the new flight details to be added in to the list
    and adds the flight details in to the CSV file

    '''
    flights = load_flights()

    flight_id = random.randint(1000, 9999)
    departure = input(f"Enter Departure City {CITIES}: ").strip()
    if departure not in CITIES:
        print("Invalid departure city.")
        return

    destination = input(f"Enter Destination City {CITIES} (must be different from departure): ").strip()
    if destination not in CITIES or destination == departure:
        print("Invalid destination city.")
        return

    try:
        price = float(input("Enter Ticket Price: "))
    except ValueError:
        print("Invalid price format.")
        return

    flights.append({
        "Flight ID": flight_id,
        "Departure": departure,
        "Destination": destination,
        "Price": price,
        "Available Seats": SEAT_CAPACITY
    })

    save_flights(flights)
    print(f"Flight {flight_id} added successfully.")


def find_flights():
    '''
    This function is used to search the flights based on the departure and destination cities.
    '''
    flights = load_flights()
    departure = input("Enter Departure City: ").strip()
    destination = input("Enter Destination City: ").strip()

    results = [flight for flight in flights if flight["Departure"] == departure and flight["Destination"] == destination]

    if results:
        print("\nAvailable Flights:")
        for flight in results:
            print(f"Flight {flight['Flight ID']} - {flight['Departure']} to {flight['Destination']}, Price: ${flight['Price']}, Seats Available: {flight['Available Seats']}")
    else:
        print("No flights found for this route.")

# Book a flight seat
def book_seat():
    '''
    This flighst takes the fight id as the input and if the flight exists in the csv file
    then flight will be booked and the available seats will be reduced by 1
    if the flight is full user will be informed about the unavailability
    '''
    flights = load_flights()
    flight_id = input("Enter Flight ID to book a seat: ").strip()

    for flight in flights:
        if flight["Flight ID"] == flight_id:
            if int(flight["Available Seats"]) > 0:
                flight["Available Seats"] = str(int(flight["Available Seats"]) - 1)
                save_flights(flights)
                print("Seat booked successfully!")
                return
            else:
                print("No seats available for this flight.")
                return

    print("Flight ID not found.")

# Cancel a booking
def cancel_booking():
    '''
    if user requests for a cancellation by entering the flight id
    then the avialable seats will be increased by 1 and booking will be cancelled
    '''
    flights = load_flights()
    flight_id = input("Enter Flight ID to cancel booking: ").strip()

    for flight in flights:
        if flight["Flight ID"] == flight_id:
            if int(flight["Available Seats"]) < SEAT_CAPACITY:
                flight["Available Seats"] = str(int(flight["Available Seats"]) + 1)
                save_flights(flights)
                print("Booking canceled successfully.")
                return
            else:
                print("All seats are already available, no bookings to cancel.")
                return

    print("Flight ID not found.")

# Generate report of flights and bookings
def generate_report():
    '''
    This function is used to generate the statastics report of the flight system
    like total available flights, total revenue generated, bookings etc.
    '''
    flights = load_flights()

    total_flights = len(flights)
    total_seats = total_flights * SEAT_CAPACITY
    booked_seats = sum(SEAT_CAPACITY - int(flight["Available Seats"]) for flight in flights)
    revenue = sum((SEAT_CAPACITY - int(flight["Available Seats"])) * float(flight["Price"]) for flight in flights)

    print("\n=== Flight Booking System Report ===")
    print(f"Total Flights: {total_flights}")
    print(f"Total Seats Available: {total_seats}")
    print(f"Seats Booked: {booked_seats}")
    print(f"Total Revenue: ${revenue:.2f}")

# Main menu function
def main():
    '''
    This is the main function that gives list of numeric values and  asks for the operation to be performed
    Based on the selection the operation will be performed.
    '''
    initialize_flight_system()

    while True:
        print("\nFlight Booking System")
        print("1. Add Flight")
        print("2. Find Flights")
        print("3. Book Seat")
        print("4. Cancel Booking")
        print("5. Generate Report")
        print("6. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == "1":
            add_flight()
        elif choice == "2":
            find_flights()
        elif choice == "3":
            book_seat()
        elif choice == "4":
            cancel_booking()
        elif choice == "5":
            generate_report()
        elif choice == "6":
            print("Exiting... Thank you for using the Flight Booking System!")
            break
        else:
            print("Invalid choice. Try again.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
import random
import csv
import os
import datetime
import string

# Constants
CSV_FILE = "flights.csv"
SEAT_CAPACITY = 100  # Each flight has 100 seats
FLIGHT_COUNT = 50

# Sample Cities for Flights
CITIES = ["New York", "Los Angeles", "Chicago", "Dallas", "San Francisco",
          "Miami", "Seattle", "Denver", "Boston", "Houston"]


def generate_flight_id():
    #generate alphanumeric flight id's of length 8
    return ''.join(random.choices(string.ascii_uppercase + string.digits, k=8))

# Validate flight ID format
def is_valid_flight_id(flight_id):
    #alpha numeric and length check for the flight id's
    return len(flight_id) == 8 and flight_id.isalnum()

# Validate date format and ensure it is not in the past
def is_valid_future_date(date_str):
    '''we are making sure that the flight date is the valid future date
    ensures the flight timing is not completed

    Raises value error if invalid dates are entered
    '''
    try:
        flight_date = datetime.datetime.strptime(date_str, "%Y-%m-%d %H:%M")
        return flight_date > datetime.datetime.now()
    except ValueError:
        return False

# Generate flights

def generate_flights(num_flights=FLIGHT_COUNT):
    '''
    The flight data is getting generated using random module
    additionally, date and time are mentioned now for the flights
    Alphanumeric 8 character flight id's are being generated
    '''
    flights = []
    for _ in range(num_flights):
        flight_id = generate_flight_id()
        departure = random.choice(CITIES)
        destination = random.choice([city for city in CITIES if city != departure])
        price = round(random.uniform(100, 1000), 2)
        available_seats = SEAT_CAPACITY
        date = (datetime.datetime.now() + datetime.timedelta(days=random.randint(1, 30))).strftime("%Y-%m-%d %H:%M")
        flights.append([flight_id, departure, destination, price, available_seats, date])
    return flights

# Initialize flight system
def initialize_flight_system():
    if not os.path.exists(CSV_FILE):
        flights = generate_flights()
        with open(CSV_FILE, "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["Flight ID", "Departure", "Destination", "Price", "Available Seats", "Date"])
            writer.writerows(flights)
        print("Flight system initialized with 50 random flights.")

# Load flights from CSV
def load_flights():
    if not os.path.exists(CSV_FILE):
        initialize_flight_system()
    with open(CSV_FILE, "r") as file:
        reader = csv.DictReader(file)
        return list(reader)

# Save flights to CSV
def save_flights(flights):
    with open(CSV_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Flight ID", "Departure", "Destination", "Price", "Available Seats", "Date"])
        for flight in flights:
          writer.writerow([flight["Flight ID"], flight["Departure"], flight["Destination"], flight["Price"], flight["Available Seats"], flight["Date"]])

# Add a new flight
def add_flight():
    '''
    Here the user enetered flight is being checked whether it is alphanumeric
    the flight id should be of length 8
    the flight id should be same as the existing flight id
    the flight id being taken from user
    '''
    flights = load_flights()

    flight_id = input("Enter Flight ID (8 alphanumeric characters): ").strip().upper()
    if not is_valid_flight_id(flight_id) or any(f["Flight ID"] == flight_id for f in flights):
        print("Invalid or duplicate Flight ID.")
        return

    departure = input(f"Enter Departure City {CITIES}: ").strip()
    if departure not in CITIES:
        print("Invalid departure city.")
        return

    destination = input(f"Enter Destination City {CITIES} (must be different from departure): ").strip()
    if destination not in CITIES or destination == departure:
        print("Invalid destination city.")
        return

    date = input("Enter flight date and time (YYYY-MM-DD HH:MM): ").strip()
    if not is_valid_future_date(date):
        print("Invalid or past date.")
        return

    '''
    Here we are handling the negative prices

    Raises value error if negatve price ie entered
    '''

    try:
        price = float(input("Enter Ticket Price: "))
        if price < 0:
            raise ValueError("Price cannot be negative.")
    except ValueError:
        print("Invalid price format.")
        return

    flights.append({
        "Flight ID": flight_id,
        "Departure": departure,
        "Destination": destination,
        "Price": price,
        "Available Seats": SEAT_CAPACITY,
        "Date": date
    })
    save_flights(flights)
    print(f"Flight {flight_id} added successfully.")

# Find flights
def find_flights():
    flights = load_flights()
    departure = input("Enter Departure City: ").strip()
    destination = input("Enter Destination City: ").strip()

    results = [flight for flight in flights if flight["Departure"] == departure and flight["Destination"] == destination]
    if results:
        for flight in results:
            print(f"Flight {flight['Flight ID']} - {flight['Departure']} to {flight['Destination']}, Price: ${flight['Price']}, Seats Available: {flight['Available Seats']}, Date: {flight['Date']}")
    else:
        print("No flights found for this route.")

# Book seat
def book_seat():
    flights = load_flights()
    flight_id = input("Enter Flight ID to book a seat: ").strip().upper()
    for flight in flights:
        if flight["Flight ID"] == flight_id:
            if int(flight["Available Seats"]) > 0:
                flight["Available Seats"] = str(int(flight["Available Seats"]) - 1)
                save_flights(flights)
                print("Seat booked successfully!")
                return
            else:
                print("No seats available for this flight.")
                return
    print("Flight ID not found.")

# Cancel booking
def cancel_booking():
    flights = load_flights()
    flight_id = input("Enter Flight ID to cancel booking: ").strip().upper()
    for flight in flights:
        if flight["Flight ID"] == flight_id:
            if int(flight["Available Seats"]) < SEAT_CAPACITY:
                flight["Available Seats"] = str(int(flight["Available Seats"]) + 1)
                save_flights(flights)
                print("Booking canceled successfully.")
                return
            else:
                print("All seats are already available.")
                return
    print("Flight ID not found.")

# Generate report
def generate_report():
    flights = load_flights()
    booked_seats = sum(SEAT_CAPACITY - int(f["Available Seats"]) for f in flights)
    revenue = sum((SEAT_CAPACITY - int(f["Available Seats"])) * float(f["Price"]) for f in flights)
    print(f"Total Flights: {len(flights)}, Seats Booked: {booked_seats}, Total Revenue: ${revenue:.2f}")

# Main menu
def main():
    initialize_flight_system()
    while True:
        choice = input("1. Add Flight 2. Find Flights 3. Book Seat 4. Cancel Booking 5. Generate Report 6. Exit: ").strip()
        if choice == "1": add_flight()
        elif choice == "2": find_flights()
        elif choice == "3": book_seat()
        elif choice == "4": cancel_booking()
        elif choice == "5": generate_report()
        elif choice == "6": break
        else: print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
